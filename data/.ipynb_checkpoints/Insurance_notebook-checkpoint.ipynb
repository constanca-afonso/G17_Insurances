```python
import pandas as pd
```

```python
df = pd.read_csv("dataexcel.csv", delimiter=";")
```

```python
df.head(3)
```

```python
Cliente = df.loc[:,['clients_id','clients_name', 'email', 'risk_level']]
Cliente.columns=['id', 'nome', 'email', 'nivel_risco']
Cliente.drop_duplicates(subset='id', inplace = True)
from sqlalchemy import create_engine
engine = create_engine('sqlite:///sqlclassestabela.db')
Cliente.to_sql('Cliente', con = engine, if_exists='replace', index= False)
```

```python
Cliente.head(3)
```

```python
Agente = df.loc[:,['agents_id','agents_name','agents_region']]
Agente.columns = ['id','nome', 'regiao']
Agente.drop_duplicates(subset='id', inplace = True)
from sqlalchemy import create_engine
engine = create_engine('sqlite:///sqlclassestabela.db')
Agente.to_sql('Agente', con = engine, if_exists='replace', index= False)
```

```python
Agente.head(3)
```

```python
nova_df = df.copy()
nova_df['id_Tipo_Apolice'] = nova_df['type'].map(lambda x: nova_df.index[nova_df['type'] == x][0])
```

```python
nova_df.head(5)
```

```python
Tipo_Apolice = nova_df.loc[:,['id_Tipo_Apolice','type']]
Tipo_Apolice.columns = ['id','designacao']
Tipo_Apolice.drop_duplicates(subset='id', inplace = True)
from sqlalchemy import create_engine
engine = create_engine('sqlite:///sqlclassestabela.db')
Tipo_Apolice.to_sql('Tipo_Apolice', con = engine, if_exists='replace', index= False)
```

```python
Tipo_Apolice.head(5)
```

```python
Apolice = nova_df.loc[:,['policies_id','coverage_amount','premium','id_Tipo_Apolice', 'agents_id']]
Apolice.columns = ['id','valor_cobertura','premio','id_Tipo_Apolice','id_agente']
Apolice.drop_duplicates(subset='id', inplace = True)
from sqlalchemy import create_engine
engine = create_engine('sqlite:///sqlclassestabela.db')
Apolice.to_sql('Apolice', con = engine, if_exists='replace', index= False)
```

```python
Apolice.head(5)
```

```python
df_Apolice_Cliente = df.loc[:,['clients_id','policies_id','effective_date','status']]
df_Apolice_Cliente['id'] = df_Apolice_Cliente.index
Apolice_Cliente = df_Apolice_Cliente.loc[:,['id','policies_id','clients_id','effective_date','status']]
Apolice_Cliente.columns = ['id','id_Apolice', 'id_Cliente', 'data_efetiva','estado']
Apolice_Cliente.drop_duplicates(subset='id', inplace = True)
from sqlalchemy import create_engine
engine = create_engine('sqlite:///sqlclassestabela.db')
Apolice_Cliente.to_sql('Apolice_Cliente', con = engine, if_exists='replace', index= False)
```
